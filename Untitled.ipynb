```python
# Mapeamento de substituição
substituicoes = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8, 'I': 9, 'J': 10, 'K': 11, 'L': 12, 'M': 13, 'N': 14, 'O': 15, 'P': 16, 'Q': 17}

# Colunas que começam com "Q0"
colunas_q0 = [col for col in df.columns if col.startswith('Q0') and col != 'Q005']

# Aplicando a substituição de forma segura
for col in colunas_q0:
    df.loc[:, col] = df[col].map(substituicoes)
```

```python
le_sg_uf_prova = LabelEncoder()
le_tp_sexo = LabelEncoder()
df.loc[:, 'SG_UF_PROVA'] = le_sg_uf_prova.fit_transform(df['SG_UF_PROVA'])
df.loc[:, 'TP_SEXO'] = le_tp_sexo.fit_transform(df['TP_SEXO'])
```

```python
corr_matrix = df.corr()

# Criando uma máscara para a parte superior da matriz
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Criando o gráfico de calor com a máscara
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm')

# Adicionando títulos e rótulos
plt.title('Matriz de Correlação com Máscara')
plt.xlabel('Variáveis do DataFrame')
plt.ylabel('Variáveis do DataFrame')

# Mostrando o gráfico
plt.show()
```

```python
corr_matrix = df.corr()
correlacao_abs_faltou = corr_matrix['Faltou'].abs()
correlacoes_ordenadas = correlacao_abs_faltou.sort_values(ascending=False)
correlacoes_ordenadas
```

```python
# Lista de variáveis com correlação maior que 10%
variaveis = ['TP_FAIXA_ETARIA', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU', 'IN_TREINEIRO', 
             'Q002', 'Q006', 'Q008', 'Q022', 'Q024', 'Q010', 'TP_ESCOLA', 'Q009', 'Q019']

# Preparando os dados para o modelo
X = df[variaveis]  # Variáveis independentes
y = df['Faltou']   # Variável dependente

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=1203)

modelo = RandomForestClassifier(random_state=1203)
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
precisao = accuracy_score(y_test, y_pred)

print(f"Precisão do Modelo: {precisao:.2f}")
```

```python
matriz_confusao = confusion_matrix(y_test, y_pred)

# Convertendo a matriz de confusão para DataFrame para melhor visualização
matriz_confusao_df = pd.DataFrame(matriz_confusao)

#Matriz de confusão
matriz_confusao_df
```

```python
prob_pred = modelo.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, prob_pred)
fpr, tpr, _ = roc_curve(y_test, prob_pred)

plt.plot(fpr, tpr, label=f"AUC: {auc:.2f}")
plt.title('Curva ROC')
plt.xlabel('Taxa de Falso Positivo')
plt.ylabel('Taxa de Verdadeiro Positivo')
plt.legend()
plt.show()
```
